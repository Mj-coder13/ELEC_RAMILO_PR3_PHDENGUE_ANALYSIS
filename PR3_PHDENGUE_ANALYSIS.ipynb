<h2>Question 1: Which year recorded the highest number of dengue cases in the Philippines?</h2>
<hr>

```python
plt.figure(figsize=(10,5))
sns.barplot(data=dengue, x='Year', y='Dengue_Cases', estimator='sum', palette='coolwarm')

plt.title("Total Dengue Cases per Year in the Philippines", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Total Dengue Cases")
plt.xticks(rotation=45)
plt.show()
```

<h2>Insight 1:</h2>
<h4>As we can see in the graph, 2019 really stands out because it had the most dengue cases â€” over 400,000 in total, which is way higher than any other year.</h4>
<HR>
<h2>Insight 2:</h2>
<h4>After 2019, the number of cases dropped a lot. This might be because of stronger awareness and better dengue prevention efforts after that big outbreak.</h4>
<HR>

<h2>Question 2: Which regions have the most dengue deaths overall?</h2>
<hr>

```python
deaths_per_region = dengue.groupby("Region")["Dengue_Deaths"].sum().sort_values(ascending=True)

plt.figure(figsize=(10,6))
plt.barh(deaths_per_region.index, deaths_per_region.values, color='lightcoral')
plt.title("Total Dengue Deaths by Region", fontsize=14, fontweight='bold')
plt.xlabel("Total Dengue Deaths")
plt.ylabel("Region")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()

deaths_per_region
```

<h2>Insight 3:</h2>
<h4>If we take a look at the chart, The National Capital Region (NCR) really stands out since it has the most dengue deaths among all regions. Itâ€™s followed by Region XII and Region VI, which also have big populations.</h4>
<hr>
<h2>Insight 4:</h2>
<h4>And the lowest numbers came from Region IV-B and Region I, suggesting that the region may have more effective dengue prevention measures, lower population density, or fewer mosquito breeding sites compared to other regions.</h4>
<hr>

<h2>Question 3: What months show the highest average dengue cases across all years?</h2>
<hr>

```python
avg_month = dengue.groupby("Month")["Dengue_Cases"].mean().reindex([
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"])

plt.figure(figsize=(12,6))
sns.lineplot(x=avg_month.index, y=avg_month.values, marker='o', color='teal')
plt.title("Average Monthly Dengue Cases (All Years)", fontsize=14, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Average Dengue Cases")
plt.xticks(rotation=30)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

<h2>Insight 5:</h2>
<h4>The line started to rise after the months of May to August, and August being the highest and I think the reason is that it may be mostly the rainy season, so there is more water and this has caused the increase in mosquitoes.</h4>
<hr>
<h2>Insight 6:</h2>
<h4>And based on the graph, April and May have the lowest averages, and theyâ€™re almost the same. Maybe because those months are usually hot and dry, so mosquitoes donâ€™t multiply that much, thatâ€™s why there are fewer dengue cases.</h4>
<hr>

<h2>Question 4: What trends can be observed between dengue cases and deaths per year?</h2>
<hr>

```python
yearly_data = dengue.groupby("Year")[["Dengue_Cases", "Dengue_Deaths"]].sum()
normalized = yearly_data / yearly_data.max()

plt.figure(figsize=(12,6))
sns.heatmap(normalized, annot=yearly_data, fmt=".0f", cmap="YlOrRd")
plt.title("Yearly Trends of Dengue Cases and Dengue Deaths", fontsize=14, fontweight='bold')
plt.xlabel("Data Type", fontsize=12)
plt.ylabel("Year", fontsize=12)
plt.tight_layout()
plt.show()
```

<h2>Insight 7:</h2>
<h4>From the heatmap, itâ€™s clear that years with more dengue cases donâ€™t always have more deaths. For example, 2019 had the highest number of cases, but 2016 had the most deaths.</h4>
<hr>
<h2>Insight 8:</h2>
<h4>This shows that having more cases doesnâ€™t always mean more people will die, which might mean that healthcare responses and early treatments got better in the later years.</h4>
<hr>

<h2>Question 5: How has the number of dengue deaths changed over time per region?</h2>
<hr>

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=dengue, x='Year', y='Dengue_Deaths', hue='Region', s=100, palette='Set1')

plt.title("Dengue Deaths Over Time by Region", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Number of Dengue Deaths")
plt.xticks(dengue['Year'].unique())
plt.grid(True, linestyle='-')
plt.show()

dengue.groupby('Year')['Dengue_Deaths'].sum().sort_values(ascending=False)
```

<h2>Insight 9:</h2>
<hr>
<h4>By looking at the scatterplot, it shows that the total number of dengue deaths decreases from 2016 to 2020, starting at 8127 deaths and falling to 1195 deaths. This downward trend is visible across almost all regions, suggesting a general improvement in dengue control over the years. Despite the overall decline, the visual emphasizes the importance of continuing preventive measures to maintain this positive trend.</h4>
<hr>
<h2>Insight 10:</h2>
<h4>Overall, dengue deaths have been decreasing over the years. However, some regions continue to report higher deaths than others. Even in years when overall deaths are lower, these regions can still have high numbers, showing that they are more affected by dengue. This also suggests that local outbreaks may happen because of factors like weather, mosquito abundance, or population vulnerability.</h4>
